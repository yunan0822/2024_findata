```python
A_1201 = (df1.loc['1201 WEICHUAN','Total current assets'] - df1.loc['1201 WEICHUAN','Total current liabilities'])/df1.loc['1201 WEICHUAN','Total assets']
```

```python
A_1201
```

```python
X1 ='Total current assets'
X2 ='Total current liabilities'
X3 ='Total assets'
X4 ='Total retained earnings'
X5 ='Earnings before interest and tax'
X6 ='Total equity'
X7 ='Total liabilities'
X8 ='Total operating revenue'
WEICHUAN ='1201 WEICHUAN'
UniPresident ='1216 UniPresident'
```

```python
A_WEICHUAN = (df1.loc[WEICHUAN,X1] - df1.loc[WEICHUAN,X2])/df1.loc[WEICHUAN, X3]
```

```python
B_WEICHUAN = df1.loc[WEICHUAN,X4]/df1.loc[WEICHUAN, X3]
```

```python
C_WEICHUAN = df2.loc[WEICHUAN,X5]/df1.loc[WEICHUAN, X3]
```

```python
D_WEICHUAN = df1.loc[WEICHUAN,X6]/df1.loc[WEICHUAN, X7]
```

```python
E_WEICHUAN = df2.loc[WEICHUAN,X8]/df1.loc[WEICHUAN, X3]
```

```python
Z_score_WEICHUAN = 1.2*A_WEICHUAN + 1.4*B_WEICHUAN + 3.3*C_WEICHUAN + 0.6*D_WEICHUAN +1.0*E_WEICHUAN
```

```python
Z_score_WEICHUAN, A_WEICHUAN, B_WEICHUAN, C_WEICHUAN, D_WEICHUAN, E_WEICHUAN
```

```python
A_UniPresident = (df1.loc[UniPresident,X1] - df1.loc[UniPresident,X2])/df1.loc[UniPresident, X3]
```

```python
B_UniPresident = df1.loc[UniPresident,X4]/df1.loc[UniPresident, X3]
```

```python
C_UniPresident = df2.loc[UniPresident,X5]/df1.loc[UniPresident, X3]
```

```python
D_UniPresident = df1.loc[UniPresident,X6]/df1.loc[UniPresident, X7]
```

```python
E_UniPresident = df2.loc[UniPresident,X8]/df1.loc[UniPresident, X3]
```

```python
Z_score_UniPresident = 1.2*A_UniPresident + 1.4*B_UniPresident + 3.3*C_UniPresident + 0.6*D_UniPresident +1.0*E_UniPresident
```

```python
Z_score_UniPresident, A_UniPresident, B_UniPresident, C_UniPresident, D_UniPresident, E_UniPresident
```
